```python
df1 = pd.read_csv("handlandmark_wide_01.csv")
df2 = pd.read_csv("handlandmark_wide_02.csv")
df3 = pd.read_csv("handlandmark_wide_03.csv")
df3_2 = pd.read_csv("handlandmark_wide_03-2.csv")
df4 = pd.read_csv("handlandmark_wide_04.csv")
df4_2 = pd.read_csv("handlandmark_wide_04-2.csv")
df6 = pd.read_csv("handlandmark_wide_06.csv")
df7 = pd.read_csv("handlandmark_wide_07.csv")
df9 = pd.read_csv("handlandmark_wide_09.csv")
df11_2 = pd.read_csv("handlandmark_wide_11-2.csv")
df1_t_1 = pd.read_csv("handlandmark_wide_01-test-01.csv")
df1_t_2 = pd.read_csv("handlandmark_wide_01-test-02.csv")
df2_t_2 = pd.read_csv("handlandmark_wide_02-test-02.csv")

print("Shape:", df1.shape)
print("Shape:", df2.shape)
print("Shape:", df3.shape)
print("Shape:", df3_2.shape)
print("Shape:", df4.shape)
print("Shape:", df4_2.shape)
print("Shape:", df6.shape)
print("Shape:", df7.shape)
print("Shape:", df9.shape)
print("Shape:", df11_2.shape)
print("Shape:", df1_t_1.shape)
print("Shape:", df1_t_2.shape)
print("Shape:", df2_t_2.shape)
```

```python
df_list = [df1, df2, df3, df3_2, df4, df4_2, df6, df7, df9, df11_2, df1_t_1, df1_t_2, df2_t_2]

gestures = ['lift', 'hold', 'press', 'adjust', 'rotate', 'place', 'other']

all_counts = {}

for i, df in enumerate(df_list, start=1):
    counts = df['behavior'].value_counts()
    result = {gesture: counts.get(gesture, 0) for gesture in gestures}
    result['total'] = sum(result.values())  
    all_counts[f'df{i}'] = result

summary_df = pd.DataFrame(all_counts).T
summary_df
```
